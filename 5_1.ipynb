```python

import torch
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification, AutoModelForQuestionAnswering, AutoModelForSeq2SeqLM
from datasets import load_dataset
import warnings
warnings.filterwarnings("ignore")

print("Готовимся к работе с русскоязычными моделями Hugging Face\n")

# ==================================================
# 1. Другая модель для анализа тональности (русский)
# ==================================================
# Одна из лучших открытых моделей 2024–2025 для русского sentiment-analysis
# https://huggingface.co/seara/rubert-tiny-sentiment-rusentiment
# (очень быстрая и точная на отзывах/новостях)

sentiment_model = "seara/rubert-tiny-sentiment-rusentiment"

sentiment_pipe = pipeline(
    "sentiment-analysis",
    model=sentiment_model,
    tokenizer=sentiment_model,
    device=0 if torch.cuda.is_available() else -1  # GPU если есть
)

print("Модель тональности загружена:", sentiment_model)

# Примеры отзывов (фильмы + товары)
reviews_ru = [
    "Фильм просто шедевр! Слезы на глазах, актёры гениальны, музыка шикарная.",
    "Очень разочарован. Сюжет предсказуемый, спецэффекты на уровне 2005 года.",
    "Телефон огонь! Камера 108 МП, батарея два дня, всё летает.",
    "Заказал наушники — пришли битые, микрофон не работает, деньги вернули с трудом."
]

print("\nЗадание 2 — Анализ тональности русских отзывов\n")
for text in reviews_ru:
    result = sentiment_pipe(text)[0]
    label = result['label']
    score = result['score']
    print(f"Отзыв: {text}")
    print(f"→ {label} (уверенность {score:.4f})\n")
```
